```python
import copy

class State_8queen:
    
    def __init__(self, board, heuristic_value=0):
        self.board = board
        # self.moves = moves
        self.heuristic_value = heuristic_value        
    
    # 새로운 상태 반환
    def get_new_board(self, i, j):

        results = []
        

        for k in range(len(self.board)):
            for l in range(len(self.board[0])):
                if self.board[k][l] == 0:
                    new_board = copy.deepcopy(self.board)
                    new_board[i][j] = 0     # 가져온 포지션에 있는 말 들어올리기
                    new_board[k][l] = 1     # 전에 들어올린 말 해당 포지션에 놓기
                    state_temp = State_8queen(new_board)
                    state_temp.heuristic_value = state_temp.h()
                    results.append(state_temp)

        return results

    # 자식 노드 생성
    def expand(self):
        results = []
        # 해당 노드의 다음 상태로 가능한 모든 경우의 상태 생성 (총 8*7=56가지)
        for i in range(8):
            for j in range(8):
                if self.board[i][j] == 1:   # 해당 포지션에 말이 존재하면 continue
                    result = self.get_new_board(i, j)
                    results += result                
                
        print(len(result))
        print(len(results))
        return results
    

    # 해당 노드의 발견적 비용 함수 계산
    def h(self):
        h_score = 0

        # 일단 복사 먼저
        new_board = copy.deepcopy(self.board)

        # 먼저 행별로 말을 고르고
        for i in range(len(new_board)):
            for j in range(len(new_board[i])):
                if new_board[i][j] == 1:
                    new_board[i][j] = 0     # 해당 말 먼저 제거

                    # 해당 말 기준으로 가로
                    for k in range(len(new_board)):
                        if new_board[i][k] == 1:
                            h_score += 1

                    # 세로
                    for k in range(len(new_board)):
                        if new_board[k][j] == 1:
                            h_score += 1
                        
                    # 왼쪽 위-아래 대각선
                    i_temp = i
                    j_temp = j
                    for k in range(len(new_board)): # 오른쪽 아래
                        i_temp += 1
                        j_temp += 1

                        if i_temp < len(new_board) and j_temp < len(new_board):  
                            if new_board[i_temp][j_temp] == 1:
                                h_score += 1
                        else:
                            break

                    i_temp = i
                    j_temp = j
                    for k in range(len(new_board)): # 왼쪽 위
                        i_temp -= 1
                        j_temp -= 1

                        if i_temp >= 0 and j_temp >= 0:  # 오른쪽 아래
                            if new_board[i_temp][j_temp] == 1:
                                h_score += 1
                        else:
                            break

                    # 왼쪽 아래-위 대각선
                    i_temp = i
                    j_temp = j
                    for k in range(len(new_board)): # 오른쪽 위
                        i_temp -= 1
                        j_temp += 1

                        if i_temp >= 0 and j_temp < len(new_board):  
                            if new_board[i_temp][j_temp] == 1:
                                h_score += 1
                        else:
                            break

                    i_temp = i
                    j_temp = j
                    for k in range(len(new_board)): # 왼쪽 아래
                        i_temp += 1
                        j_temp -= 1

                        if i_temp < len(new_board) and j_temp >= 0:  
                            if new_board[i_temp][j_temp] == 1:
                                h_score += 1
                        else:
                            break
                        
            #     print(h_score)
            # print(new_board)

        self.h = h_score
        return h_score


    # 노드 상태 출력
    def __str__(self):
        return "h:" + str(self.heuristic_value) + "\n" + \
        str(self.board[0]) + "\n" + \
        str(self.board[1]) + "\n" + \
        str(self.board[2]) + "\n" + \
        str(self.board[3]) + "\n" + \
        str(self.board[4]) + "\n" + \
        str(self.board[5]) + "\n" + \
        str(self.board[6]) + "\n" + \
        str(self.board[7]) + "\n" + \
        "-------------------------"
```
